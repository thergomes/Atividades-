Esther de Oliveira Gomes 

```python
import numpy as np
import matplotlib.pyplot as plt

#a)
t = np.arange (0,5,0.1)
h = -t**2 + 4*t

plt.figure()
plt.plot(t, h)
plt.xlabel('tempo(s)')
plt.ylabel('altura(m)')
plt.title('Gráfico da altura em função do tempo')
plt.show()

#b)
hmax = np.max(h)
print(hmax)
```
